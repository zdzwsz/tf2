```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=1000
x_data = np.linspace(0,10,N)[:,np.newaxis]
noise = np.random.normal(0.0,0.02,(N,1))
y_data = 5 * x_data + 4 + noise

y_data1 = x_data * x_data + noise
#x = tf.random.normal([2, 3])
#print(x_data.shape)
#print(x_data)
#print(y_data)
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#model = tf.keras.Sequential([               #   定义全连接层结构
#    tf.keras.layers.Dense(4,activation='relu'),                 #   第一层输出为4
#    tf.keras.layers.Dense(3,activation='relu'),
#    tf.keras.layers.Dense(2)                                    #   输出层不需要激活函数
#])

model.build(input_shape=(1,1))#模型的搭建
model.summary()#查看每一层的信息
```

```python
model.compile(loss='mse',optimizer='sgd')
model.fit(x_data, y_data ,verbose=1,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

```python
model1 = tf.keras.Sequential([               #   定义全连接层结构
   tf.keras.layers.Dense(128,activation='relu'),     
   tf.keras.layers.Dense(64, activation='relu'),
   tf.keras.layers.Dense(1)
])
model1.build(input_shape=(1,1))#模型的搭建
model1.summary()#查看每一层的信息
```

```python
model1.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(lr=0.0001))
model1.fit(x_data, y_data1 ,verbose=1,epochs=2000,validation_split=0.1)
```

```python
x1 = tf.constant([2])
y1 = model1.predict(x1)
print(y1)
```
